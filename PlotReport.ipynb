## Bar plots για συνδυασμους τιμων H1, H2 για την tanh, relu με batch size = 1

```python
fig, ax = plt.subplots(1, 2, figsize = (21, 7))

df1 = pd.DataFrame({'MLP':['H1=2, H2=1', 'H1=3, H2=2', 'H1=4, H2=3', 'H1=5, H2=4', 'H1=6, H2=5'],
                    'gen-ability':[44.12, 53.99, 79.72, 81.62, 86.02]})
ax[0] = df1.plot.bar(x = 'MLP', y = 'gen-ability', ax = ax[0])
ax[0].title.set_text('Generalization Ability (Tanh) vs different H1, H2 numbers, Batch Size = 1')
ax[0].set_ylabel('max generalization ability')

df2 = pd.DataFrame({'MLP':['H1=2, H2=1', 'H1=3, H2=2', 'H1=4, H2=3', 'H1=5, H2=4', 'H1=6, H2=5'],
                    'gen-ability':[23.08, 50.92, 74.6, 77.96, 36.17]})
df2.plot.bar(x = 'MLP', y = 'gen-ability', color = 'orange', ax = ax[1])
ax[1].title.set_text('Generalization Ability (Relu) vs different H1, H2 numbers, Batch Size = 1')
ax[1].set_ylabel('max generalization ability')
```
